# Importação de bibliotecas

```python
import pandas as pd
import numpy as np
import sklearn.preprocessing
```

# Coleta de Dados

```python
# Dimensoes

df_cliente = pd.read_csv('tabela_cliente.csv', encoding='latin-1')
df_funcionario = pd.read_csv('tabela_funcionario.csv', encoding='latin-1')
df_departamento = pd.read_csv('tabela_departamento.csv', encoding='latin-1')
df_projeto = pd.read_csv('tabela_projeto.csv', encoding='latin-1')
df_func_proj = pd.read_csv('tabela_funcionario_projeto.csv', encoding='latin-1')
```

# Tratamento de dados

```python
# from sklearn.preprocessing import LabelEnconder

LE = sklearn.preprocessing.LabelEncoder()
df_funcionario['supervisor_number'] = LE.fit_transform(df_funcionario['supervisor'])
```

# Analise dos dados
